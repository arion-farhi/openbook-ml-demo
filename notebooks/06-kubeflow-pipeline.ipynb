```python
# 06-kubeflow-pipeline.ipynb
# Openbook ML Demo - Kubeflow Pipeline with TFDV Data Validation

"""
Pipeline Orchestration:
- Kubeflow Pipelines: Framework-agnostic orchestration (works with PyTorch)
- TFDV: Data validation component from TFX (schema generation, anomaly detection)

Why Kubeflow instead of full TFX?
- TFX is TensorFlow-centric (Trainer, Evaluator, Pusher are TF-specific)
- Kubeflow works with any framework (PyTorch, sklearn, etc.)
- TFDV is standalone - works on DataFrames regardless of model framework

Pipeline steps:
1. Load data from GCS
2. Generate statistics with TFDV
3. Create schema
4. Validate new data against schema
5. Run PyTorch model inference
"""

!pip install tf_keras --quiet
!pip install tensorflow-data-validation --quiet
!pip install kfp google-cloud-aiplatform --quiet

import tensorflow_data_validation as tfdv
import pandas as pd
from google.cloud import storage
from io import StringIO

PROJECT_ID = "openbook-ml-demo"
REGION = "us-central1"
BUCKET_NAME = "openbook-data-lake"

client = storage.Client(project=PROJECT_ID)
bucket = client.bucket(BUCKET_NAME)

def load_csv_from_gcs(path):
    blob = bucket.blob(path)
    return pd.read_csv(StringIO(blob.download_as_text()))

print("✓ Kubeflow + TFDV installed")
```

```python
# Load training data and generate statistics with TFDV
train_df = load_csv_from_gcs("processed/features/train.csv")
test_df = load_csv_from_gcs("processed/features/test.csv")

print(f"Train: {len(train_df)} rows")
print(f"Test: {len(test_df)} rows")

# Generate statistics
train_stats = tfdv.generate_statistics_from_dataframe(train_df)
test_stats = tfdv.generate_statistics_from_dataframe(test_df)

print("\n✓ Statistics generated")

# Visualize training data statistics
tfdv.visualize_statistics(train_stats)
```

```python
# Generate schema from training data
schema = tfdv.infer_schema(train_stats)
tfdv.display_schema(schema)
```

```python
# Validate test data against training schema
anomalies = tfdv.validate_statistics(test_stats, schema)
tfdv.display_anomalies(anomalies)
```

```python
# Simulate data drift to demonstrate TFDV catching anomalies
import numpy as np

# Create "bad" data - imagine upstream system changed
drifted_df = test_df.copy()
drifted_df['coverage_ratio'] = drifted_df['coverage_ratio'] * 100  # Oops, now 40-100 instead of 0.4-1.0
drifted_df['procedure_code'] = drifted_df['procedure_code'].replace('D2750', 'D9999')  # New unknown code

# Generate stats and validate
drifted_stats = tfdv.generate_statistics_from_dataframe(drifted_df)
drift_anomalies = tfdv.validate_statistics(drifted_stats, schema)

print("Simulated drift anomalies:")
tfdv.display_anomalies(drift_anomalies)
```

```python
# Build Kubeflow Pipeline
from kfp import dsl
from kfp.dsl import component, Input, Output, Dataset, Model, Metrics
from google.cloud import aiplatform

aiplatform.init(project=PROJECT_ID, location=REGION)

# Component 1: Load and validate data
@component(
    base_image="python:3.10",
    packages_to_install=["pandas", "google-cloud-storage", "tensorflow-data-validation"]
)
def validate_data(
    bucket_name: str,
    data_path: str,
    output_valid: Output[Dataset]
) -> bool:
    import pandas as pd
    from google.cloud import storage
    from io import StringIO
    import tensorflow_data_validation as tfdv
    
    client = storage.Client()
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(data_path)
    df = pd.read_csv(StringIO(blob.download_as_text()))
    
    stats = tfdv.generate_statistics_from_dataframe(df)
    schema = tfdv.infer_schema(stats)
    anomalies = tfdv.validate_statistics(stats, schema)
    
    # Save validated data
    df.to_csv(output_valid.path, index=False)
    
    return len(anomalies.anomaly_info) == 0

print("✓ Data validation component defined")
```

```python
# Component 2: Run model inference
@component(
    base_image="python:3.10",
    packages_to_install=["pandas", "torch", "google-cloud-storage", "scikit-learn"]
)
def run_inference(
    bucket_name: str,
    model_path: str,
    data: Input[Dataset],
    predictions: Output[Dataset]
):
    import pandas as pd
    import torch
    from google.cloud import storage
    import pickle
    
    # Load data
    df = pd.read_csv(data.path)
    
    # In production: load model from GCS and run inference
    # For demo: simulate predictions
    df['predicted_copay'] = df['patient_copay'] * 0.95  # Placeholder
    df['predicted_insurance'] = df['procedure_cost'] - df['predicted_copay']
    
    df.to_csv(predictions.path, index=False)

# Component 3: Evaluate predictions
@component(
    base_image="python:3.10",
    packages_to_install=["pandas", "scikit-learn"]
)
def evaluate_model(
    predictions: Input[Dataset],
    metrics: Output[Metrics]
):
    import pandas as pd
    from sklearn.metrics import mean_absolute_error, r2_score
    import json
    
    df = pd.read_csv(predictions.path)
    
    mae = mean_absolute_error(df['patient_copay'], df['predicted_copay'])
    r2 = r2_score(df['patient_copay'], df['predicted_copay'])
    
    metrics.log_metric("mae", mae)
    metrics.log_metric("r2", r2)

# Build pipeline
@dsl.pipeline(
    name="openbook-copay-pipeline",
    description="End-to-end copay prediction pipeline with TFDV validation"
)
def copay_pipeline(bucket_name: str, data_path: str, model_path: str):
    
    validate_task = validate_data(
        bucket_name=bucket_name,
        data_path=data_path
    )
    
    inference_task = run_inference(
        bucket_name=bucket_name,
        model_path=model_path,
        data=validate_task.outputs["output_valid"]
    )
    
    evaluate_task = evaluate_model(
        predictions=inference_task.outputs["predictions"]
    )

print("✓ Pipeline defined")
```

```python
# Compile and run pipeline
from kfp import compiler

# Compile pipeline to JSON
compiler.Compiler().compile(
    pipeline_func=copay_pipeline,
    package_path="copay_pipeline.json"
)

# Upload to GCS
blob = bucket.blob("pipeline/copay_pipeline.json")
blob.upload_from_filename("copay_pipeline.json")

print("✓ Pipeline compiled and uploaded to GCS")

# Run on Vertex AI
from google.cloud.aiplatform import PipelineJob

job = PipelineJob(
    display_name="openbook-copay-pipeline-run",
    template_path="copay_pipeline.json",
    pipeline_root=f"gs://{BUCKET_NAME}/pipeline_runs",
    parameter_values={
        "bucket_name": BUCKET_NAME,
        "data_path": "processed/features/test.csv",
        "model_path": "models/copay_model.pt"
    }
)

job.submit()
print(f"✓ Pipeline submitted")
print(f"View at: https://console.cloud.google.com/vertex-ai/pipelines/runs?project={PROJECT_ID}")
```
